```python
# Step 1: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Step 2: Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load the uploaded dataset
# This assumes the file is named exactly 'student_dropout.csv'
data = pd.read_csv('student_dropout.csv')

# Optional: View first few rows to understand the dataset structure
data.head()

# Step 4: Preprocess the data
# Check column names (in case they're different)
print("Columns in dataset:", data.columns)

# Separate features and target
X = data[['attendance', 'grades', 'participation']]  # Change if your column names differ
y = data['dropout_risk']  # The column you want to predict (ensure it's correctly named)

# Handle missing values (if any)
X.fillna(X.mean(), inplace=True)
y.fillna(y.mode()[0], inplace=True)

# Step 5: Encode the target labels (if they're in string format like 'yes'/'no')
le = LabelEncoder()

# Encode 'dropout_risk' target column
y = le.fit_transform(y)  # This will transform 'yes' -> 1, 'no' -> 0 or equivalent

# Step 6: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 7: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 8: Train the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Step 9: Evaluate the model and print metrics

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix heatmap
disp = ConfusionMatrixDisplay(conf_matrix, display_labels=le.classes_)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix - Student Dropout Prediction')
plt.show()

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print results
print(f"✅ Accuracy:  {accuracy:.2f}")
print(f"✅ Precision: {precision:.2f}")
print(f"✅ Recall:    {recall:.2f}")
print(f"✅ F1 Score:  {f1:.2f}")

# Step 10: Optional - Feature Importance (to understand the model better)
importances = model.feature_importances_
feature_names = X.columns
feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(6, 4))
sns.barplot(x=feat_imp.values, y=feat_imp.index, palette="viridis")
plt.title("Feature Importance - Student Dropout Prediction")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# Step 11: Optional - Export predictions to a CSV file
results_df = pd.DataFrame(X_test, columns=feature_names)
results_df['Actual'] = y_test
results_df['Predicted'] = y_pred
results_df.to_csv('dropout_predictions.csv', index=False)

# Allow the user to download the predictions file
from google.colab import files
files.download('dropout_predictions.csv')

```
